# Usecase 3: Absolute abundance prediction original model set-up


The scripts mentioned in this notebook train XGBoost Regressors in Python with the same train-test splits as the ritme runs and a modelling set-up equivalent to the one described in [the original publication by Nishijima et al. 2024](https://www.sciencedirect.com/science/article/pii/S0092867424012042) and [the official GitHub Repository of the publication](https://github.com/grp-bork/CellCount_Nishijima_2024/blob/main/construct_models.R).

Notes: 
* According to the [original code repos](https://github.com/grp-bork/CellCount_Nishijima_2024/blob/main/construct_models.R) of the publication the R2 was maximised not the RMSE - hence we implement this accordingly.
* [Original code repos](https://github.com/grp-bork/CellCount_Nishijima_2024/blob/main/construct_models.R) scales based on complete dataset not just train set - we train the scaler only on the train set to avoid data leakage and apply it on the test set.
* For a proper comparison with ritme, we are running inner cross-validation on the train set and evaluating the resulting model on the left-out test set.

The scripts can be run in the following conda environment:

```shell
mamba create -n ritme_v128 -c adamova -c qiime2 -c conda-forge -c bioconda -c pytorch -c anaconda ritme ipykernel -y
conda activate ritme_v128
pip install -e .
```
